# Exploratory Data Analysis (EDA) of Pakistan Census Dataset

## 1. Introduction

This project performs exploratory data analysis on a Pakistan census based dataset containing population, demographic, and administrative information across different regions of Pakistan. The dataset includes variables related to population distribution, gender breakdown, urbanization, growth rates, and administrative divisions.

The objective of this analysis is to understand population patterns across provinces, districts, and tehsils and to identify key demographic insights using statistical exploration.



#Dataset Description
Source: Kaggle – Pakistan Census 2017 Dataset
URL: ("https://www.kaggle.com/datasets/usamaraheem/pakistan-census-2017")
Rows: 10,000+
Columns: 17
Time Period: 2017 Census
Collected by: Pakistan Bureau of Statistics (PBS)
Type: Demographic + population statistics
Unit: District/Tehsil level records

## 3. Column Description

- adm_lvl_1 → Province
- adm_lvl_2 → Division
- adm_lvl_3 → District
- adm_lvl_4 → Tehsil/Sub-division
- area_name → Name of area
- area_type → Total / Urban / Rural classification
- area_sq_km → Area in square kilometers
- total_pop → Total population
- total_male → Male population
- total_female → Female population
- total_trans → Transgender population
- sex_ratio → Males per 100 females
- pop_density → Population per sq km
- pct_urban → Percentage urban population
- avg_hh_size → Average household size
- total_pop_98 → Population in 1998
- avg_annual_growth → Annual growth rate

## 4. Data Quality Observations

- The dataset contains missing values in several columns such as area_sq_km, pop_density, and pct_urban due to hierarchical data structure.
- Some variables are only available at aggregated levels (e.g., province totals), while others are missing at rural or urban levels.
- The dataset follows a multi-level administrative structure which results in structured missing data rather than random missingness.

```python
cat_cols = df.select_dtypes(include='object').columns

for col in cat_cols:
    print("\nColumn:", col)
    print(df[col].value_counts())
```

```python
#missing values 
df.isnull().sum()
```

```python
#missing percentage 
missing_percent = (df.isnull().sum() / len(df)) * 100
missing_percent.sort_values(ascending=False)
```

```python
#OUTLIERS
num_cols = df.select_dtypes(include=np.number).columns

for col in num_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1

    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    outliers = df[(df[col] < lower) | (df[col] > upper)]

    print(f"{col}: {len(outliers)} outliers")
```

```python
#suspisicious data checking 
df.describe(include='all')
for col in num_cols:
    print(col, "negative values:", (df[col] < 0).sum())
```

```python
df.columns
```

```python
df_total = df[df["area_type"] == "TOTAL"]
```

```python
#province highest total population
df.groupby("adm_lvl_1")["total_pop"].sum().sort_values(ascending=False)
```

```python
#district  highest population
result = df[df["adm_lvl_3"].notna()].groupby("adm_lvl_3")["total_pop"].sum().sort_values(ascending=False).head(10)
print(result)
```

```python
print("Total Population of Pakistan:", df_total["total_pop"].sum())
```

```python
#province highest population density 
result = df_total.groupby("adm_lvl_1")["pop_density"].mean().sort_values(ascending=False)
print(result)
```

```python
# male vs female population 
print("Male Population:", df_total["total_male"].sum())
print("Female Population:", df_total["total_female"].sum())
```

```python
#which area type has highest population 
result = df.groupby("area_type")["total_pop"].sum()
print(result)
```
